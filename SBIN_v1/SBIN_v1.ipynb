Starting With Actual Code. Importing Pandas, Numpy, Tensorflow, keras,TvDataFeed.

```python
import pandas as pd
import numpy as np
import keras
import tensorflow as tf
print(tf.__version__)
```

TvDataFeed, a data feed from Trading View API. API has limitation with free account.



```python
from tvDatafeed import TvDatafeed, Interval
username = 'username'
password = 'password'
tv = TvDatafeed(username=username,password=password)
```

Fetch Daily Historical data. n_bars = 10000, covers all the data.

```python
df = tv.get_hist('SBIN','NSE',interval=Interval.in_daily,n_bars=10000)
df.to_csv('SBIN_DAILY.csv')
df
```

Preprocessing data from training the model.

This model is going to use only 'close' and 'volume' to keep the initial model version simple.
Slice the data to a more recent window, data since 1999 might contain outdated patterns and behaviour. So selecting since 2014-2025 data is more relevant.

```python
df = df.loc['2014-01-01 03:45:00': ,['close','volume']]
corr = df.corr()
df,corr
```

We need to scale down the data. The Scalers should be independent from both features as the close price is smaller number(eg. 800) and volume is greater number(eg, 179000).

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_close = MinMaxScaler()
scaler_volume = MinMaxScaler()

df['close'] = scaler_close.fit_transform(df[['close']]).flatten()
df['volume'] = scaler_volume.fit_transform(df[['volume']]).flatten()

df
```

Need to save the scaler for inverse scaling the prediction values.

```python
import joblib
joblib.dump(scaler_close,'scaler_close.pkl')
joblib.dump(scaler_volume,'scaler_volume.pkl')
```

Every prediction needs to lookup previous bars for trend and pattern. So about 100 bars might be sufficient for any pattern/trend recognition. There for we need 100 lagged bars(eg. t-1,t-2...t-100) and 1 present bar(eg. t) for training.

Pair \[close,volume\] makes the row a (100,2) + 1 (lag colums + target close).

We do not need volume, as the target is to only predict close price.

```python
cols = {}
for i in range(1, 101):
    cols[f'close (t-{i})'] = df['close'].shift(i)
    cols[f'volume (t-{i})'] = df['volume'].shift(i)

df = pd.concat([df] + [pd.Series(cols[col], name=col) for col in cols], axis=1)
df.pop('volume')
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)

cols = {}
for i in range(1, 101):
    cols[f'(t-{i})'] = [[x, y] for x, y in zip(df[f'close (t-{i})'], df[f'volume (t-{i})'])]

drop_columns = [f'close (t-{i})' for i in range(1, 101)] + [f'volume (t-{i})' for i in range(1, 101)] + ['volume']

df = pd.concat([df[['close']]] + [pd.Series(cols[col], name=col) for col in cols],axis=1)

df
```

Train test, Split the data 80-20 standard ratio.

```python
from sklearn.model_selection import train_test_split
y = df.pop('close').values
x = np.array([np.array(df.iloc[i].values.tolist()) for i in range(len(df))])

x,y
```

```python
x_train,x_eval = train_test_split(x,test_size=0.20,shuffle=True,random_state=42)
x_train.shape,x_eval.shape
```

```python
y_train,y_eval = train_test_split(y,test_size=0.20,shuffle=True,random_state=42)
len(y_train),len(y_eval)
```

Now we done with preprocessing the data.

We create a model now and train it.
So, for v1 let's test using single layer of LSTM with 50 units. Dropout 0.2 and Dense 25 to Dense 1(Output).

With adam optimizer and mse loss function.

Early stopping with monitor on val_loss and patience at 10, and try with restore best weights.

```python
from keras.layers import LSTM,Input, Dense, Dropout
from keras import Sequential
from keras.callbacks import EarlyStopping
```

```python
model = Sequential([
    Input(shape=(100,2)),
    LSTM(50),
    Dropout(0.2),
    Dense(25,activation='relu'),
    Dense(1)
])
model.compile(optimizer='adam',loss='mse')
early_stopping = EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
history = model.fit(x_train,y_train,epochs=100,batch_size=64,validation_split=0.1,callbacks=[early_stopping],verbose=1)
```

Lets save the model as .keras

```python
model.save("sbin_v1.keras")
```

Testing time.

```python
y_pred = model.predict(x_eval, batch_size=64)
y_pred_inr = scaler_close.inverse_transform(y_pred.reshape(-1, 1)).flatten()
y_test_inr = scaler_close.inverse_transform(y_eval.reshape(-1, 1)).flatten()
```

RMSE - 36.468010286112026

```python
from sklearn.metrics import root_mean_squared_error

rmse = root_mean_squared_error(y_test_inr, y_pred_inr)
rmse
```

Directional Accuracy: 94.53%

```python
direction_correct = np.sum((y_pred_inr[1:] > y_test_inr[:-1]) == (y_test_inr[1:] > y_test_inr[:-1]))
directional_accuracy = direction_correct / (len(y_test_inr) - 1) * 100
print(f"Directional Accuracy: {directional_accuracy:.2f}%")
```
