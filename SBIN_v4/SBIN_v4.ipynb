Fetch Daily Historical data. n_bars = 10000, covers all the data.

```python
df = tv.get_hist('SBIN','NSE',interval=Interval.in_daily,n_bars=10000)
df.to_csv('SBIN_DAILY.csv')
df
```

Preprocessing data from training the model.

This model is going to use 'close', 'volume change','volume', 'rsi' and 'ma5'.

```python
df = df.loc[: ,['close','volume']]
corr = df.corr()
df,corr
```

```python
df['volume change'] = df['volume'].shift(1).fillna(0)
df
```

```python
delta = df['close'].diff()
gain = delta.clip(lower=0).rolling(window=14).mean()
loss = -delta.clip(upper=0).rolling(window=14).mean()
rs = gain / loss
df['rsi'] = 100 - (100 / (1 + rs))
df
```

```python
df['ma5'] = df['close'].rolling(window=5).mean()
df
```

We need to scale down the data. The Scalers should be independent from both features as the close price is smaller number(eg. 800) and volume is greater number(eg, 179000).

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_close = MinMaxScaler()
scaler_volume = MinMaxScaler()
scaler_volume_change = MinMaxScaler()
scaler_rsi = MinMaxScaler()
scaler_ma5 = MinMaxScaler()

df['close'] = scaler_close.fit_transform(df[['close']]).flatten()
df['volume'] = scaler_volume.fit_transform(df[['volume']]).flatten()
df['volume change'] = scaler_volume_change.fit_transform(df[['volume change']]).flatten()
df['rsi'] = scaler_rsi.fit_transform(df[['rsi']]).flatten()
df['ma5'] = scaler_ma5.fit_transform(df[['ma5']]).flatten()

df
```

Need to save the scaler for inverse scaling the prediction values.

```python
import joblib
joblib.dump(scaler_close,'scaler_close.pkl')
joblib.dump(scaler_volume,'scaler_volume.pkl')
joblib.dump(scaler_volume_change,'scaler_volume_change.pkl')
joblib.dump(scaler_volume_change,'scaler_rsi.pkl')
joblib.dump(scaler_volume_change,'scaler_ma5.pkl')
```

```python
df = df.loc['2020-01-01 03:45:00':]
df
```

Now we are going to use 60 bars because 100 bars from previous version might be creating noise resulting in high rmse. Therefore we need 60 lagged bars(eg. t-1,t-2...t-60) and 1 present bar(eg. t) for training.

Pair \[close,volume\] makes the row a (60,5) + 1 (lag colums + target close).

We do not need volume, as the target is to only predict close price.

```python
cols = {}
for i in range(1, 61):
    cols[f'close (t-{i})'] = df['close'].shift(i)
    cols[f'volume (t-{i})'] = df['volume'].shift(i)
    cols[f'volume change (t-{i})'] = df['volume change'].shift(i)
    cols[f'rsi (t-{i})'] = df['rsi'].shift(i)
    cols[f'ma5 (t-{i})'] = df['ma5'].shift(i)

df = pd.concat([df] + [pd.Series(cols[col], name=col) for col in cols], axis=1)
df.pop('volume')
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)

cols = {}
for i in range(1, 61):
    cols[f'(t-{i})'] = [[x,y,z,a,b] for x,y,z,a,b in zip(df[f'close (t-{i})'], df[f'volume (t-{i})'], df[f'volume change (t-{i})'],df[f'rsi (t-{i})'],df[f'ma5 (t-{i})'])]

drop_columns = [f'close (t-{i})' for i in range(1, 61)] + [f'volume (t-{i})' for i in range(1, 61)] + ['volume'] + [f'volume change (t-{i})' for i in range(1, 61)] + [f'ma5 (t-{i})' for i in range(1, 61)] + [f'rsi (t-{i})' for i in range(1, 61)]

df = pd.concat([df[['close']]] + [pd.Series(cols[col], name=col) for col in cols],axis=1)

df
```

Train test, Split the data 80-20 standard ratio.

```python
from sklearn.model_selection import train_test_split
y = df.pop('close').values
x = np.array([np.array(df.iloc[i].values.tolist()) for i in range(len(df))])

x,y
```

```python
x_train,x_eval = train_test_split(x,test_size=0.20,shuffle=True,random_state=42)
x_train.shape,x_eval.shape
```

```python
y_train,y_eval = train_test_split(y,test_size=0.20,shuffle=True,random_state=42)
len(y_train),len(y_eval)
```

Now we done with preprocessing the data.

We create a model now and train it.
So, for v4 let's test using 2 layers of LSTM with 100 units and 50 units. Dropout 0.4 and Dense 25 to Dense 1(Output), with L2 regularization.

With adam optimizer and mse loss function.

Early stopping with monitor on val_loss and patience at 15, and try with restore best weights.

```python
from keras.layers import LSTM,Input, Dense, Dropout
from keras import Sequential
from keras.regularizers import l2
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

```python
model = Sequential([
    Input(shape=(100,5)),
    LSTM(100,return_sequences=True,kernel_regularizer=l2(0.01)),
    Dropout(0.4),
    LSTM(50,kernel_regularizer=l2(0.01)),
    Dropout(0.4),
    Dense(25,activation='relu',kernel_regularizer=l2(0.01)),
    Dense(1)
])
model.compile(optimizer='adam',loss='mse')
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.0001)
early_stopping = EarlyStopping(monitor='val_loss',patience=15,restore_best_weights=True)
```

```python
history = model.fit(x_train,y_train,epochs=100,batch_size=32,validation_split=0.2,callbacks=[early_stopping,lr_scheduler],verbose=1)
```

Lets save the model as .keras

```python
model.save("sbin_v4.keras")
```

Testing time.

```python
y_pred = model.predict(x_eval, batch_size=64)
y_pred_inr = scaler_close.inverse_transform(y_pred.reshape(-1, 1)).flatten()
y_test_inr = scaler_close.inverse_transform(y_eval.reshape(-1, 1)).flatten()
```

RMSE - 21.478414472894976

```python
from sklearn.metrics import root_mean_squared_error

rmse = root_mean_squared_error(y_test_inr, y_pred_inr)
rmse
```

Directional Accuracy: 96.55%

```python
direction_correct = np.sum((y_pred_inr[1:] > y_test_inr[:-1]) == (y_test_inr[1:] > y_test_inr[:-1]))
directional_accuracy = direction_correct / (len(y_test_inr) - 1) * 100
print(f"Directional Accuracy: {directional_accuracy:.2f}%")
```
